```python
import fitz
import json
from transformers import pipeline

# Use a smaller model to avoid memory errors
llm = pipeline("text-generation", model="distilgpt2", max_new_tokens=100, device=-1)

pdf_path = r"c:\Users\91885\Downloads\FINANCIAL RESULTS 1.pdf" 

def extract_text_from_pdf(pdf_path):
    doc = fitz.open(pdf_path)
    text = "\n".join(page.get_text("text") for page in doc)
    return text

def chunk_text(text, chunk_size=512):  # Limit each chunk to a size that fits model limits
    return [text[i:i + chunk_size] for i in range(0, len(text), chunk_size)]

def extract_financial_data(text, company_name):
    # Updated prompt to ensure output in desired JSON format
    prompt = f"""
    Extract the financial details for {company_name} from the text:
    The output must strictly follow this JSON format:
    {{
        "{company_name}": {{
            "company_name": "{company_name}",
            "report_date": "Extracted date",
            "profit_before_tax": {{
                "standalone": [value1, value2, value3, value4, value5],
                "consolidated": [value1, value2, value3, value4, value5]
            }}
        }}
    }}

    Text:
    {text}
    """
    
    response = llm(prompt)[0]["generated_text"]
    print("Raw model output:", response)  # Print the raw response for inspection
    
    try:
        # Try to parse the model output
        extracted_json = json.loads(response)
    except json.JSONDecodeError:
        print("⚠ Warning: LLM output was not valid JSON.")
        extracted_json = {"error": "LLM output was not valid JSON"}
    except Exception as e:
        print(f"Error occurred: {e}")
        extracted_json = {"error": str(e)}

    return extracted_json

def save_to_json(data, output_path):
    with open(output_path, "w", encoding="utf-8") as f:
        json.dump(data, f, indent=4, ensure_ascii=False)

# Extract and chunk the text
text = extract_text_from_pdf(pdf_path)
chunks = chunk_text(text)

final_output = {}

# Process each chunk of text separately
for chunk in chunks:
    print(f"Processing chunk...")
    
    # Extracting data for Eveready Industries India Ltd.
    eveready_data = extract_financial_data(chunk, "Eveready Industries India Ltd.")
    if "Eveready Industries India Ltd." in eveready_data:
        final_output.update(eveready_data)
    
    # Extracting data for Singhi & Co.
    singhi_data = extract_financial_data(chunk, "Singhi & Co.")
    if "Singhi & Co." in singhi_data:
        final_output.update(singhi_data)

# Save the output to a JSON file
save_to_json(final_output, "financial_data1.json")

print("✅ Extraction complete! Data saved to financial_data1.json")
```
