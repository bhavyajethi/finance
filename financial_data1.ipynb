```python
import fitz  # PyMuPDF for PDF extraction
import json
from transformers import pipeline

# ✅ Use Open-Source Model (No Token Needed)
llm = pipeline("text-generation", model="microsoft/phi-1_5")



# ✅ Extract text from PDF
def extract_text_from_pdf(pdf_path):
    doc = fitz.open(pdf_path)
    text = "\n".join(page.get_text("text") for page in doc)
    return text

# ✅ Extract financial data using LLM (Ensuring JSON Format)
def extract_financial_data(text, company_name):
    prompt = f"""
    Extract financial details for {company_name} from the text:
    - Company Name
    - Report Date
    - Profit Before Tax (Standalone - 5 values, Consolidated - 5 values)
    
    Provide the response **strictly in JSON format** with proper key-value pairs.

    Text:
    {text}
    """

    response = llm(prompt)[0]["generated_text"]
    
    # ✅ Ensure JSON Format
    try:
        extracted_json = json.loads(response)
    except json.JSONDecodeError:
        print("⚠ Warning: LLM output is not valid JSON. Trying to fix formatting...")
        extracted_json = {"error": "LLM output was not valid JSON"}

    return extracted_json

# ✅ Save data to JSON
def save_to_json(data, output_path):
    with open(output_path, "w", encoding="utf-8") as f:
        json.dump(data, f, indent=4, ensure_ascii=False)

# ✅ Process PDF and extract details for both companies
pdf_path = "FINANCIAL RESULTS 1.pdf"
text = extract_text_from_pdf(pdf_path)

eveready_data = extract_financial_data(text, "Eveready Industries India Ltd.")
singhi_data = extract_financial_data(text, "Singhi & Co.")

final_output = {
    "Eveready Industries India Ltd.": eveready_data,
    "Singhi & Co.": singhi_data
}

# ✅ Save final JSON output
save_to_json(final_output, "financial_data.json")

print("✅ Extraction complete! Data saved to financial_data.json")
```
